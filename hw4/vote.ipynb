```python
import pandas as pd
import numpy as np
```

```python
answer_target=[18,16,14,13,9]
```

```python
allanswer=list()
for i in answer_target:
    allanswer.append(pd.read_csv("answer/answer_"+str(i)+".csv").set_index('id'))
```

```python
vote=pd.DataFrame(data={"label":0},index=allanswer[0].index)
```

```python
for ans in allanswer:
    vote=vote+ans
```

```python
vote.label=(vote['label']>=3).astype(int)
```
